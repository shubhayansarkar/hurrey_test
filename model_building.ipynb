```python
df.price=df.groupby('variety')['price'].transform(lambda x: x.fillna(x.median()))
```

```python
df.region_1=df.groupby(['country','province'])['region_1'].transform(lambda x: x.fillna(f'others'))
```

```python
df.dropna(inplace=True)
```

```python
df.region_1.value_counts()
```

```python
df.loc[(df['points'] > 80) & (df['points'] <=85), 'wine_quality'] = 'Average'
df.loc[(df['points'] > 85) & (df['points'] <=90), 'wine_quality'] = 'Good'
df.loc[(df['points'] > 90) & (df['points'] <=95), 'wine_quality'] = 'Great'
df.loc[(df['points'] > 95) & (df['points'] <=100), 'wine_quality'] = 'Perfect'
```

```python
df.loc[(df['price'] > 0) & (df['price'] <=25) , 'price_val'] = '0-25'
df.loc[(df['price'] > 26) & (df['price'] <=50) , 'price_val'] = '26-50'
df.loc[(df['price'] > 51) & (df['price'] <=75) , 'price_val'] = '51-75'
df.loc[(df['price'] > 76) & (df['price'] <=100) , 'price_val'] = '76-100'

df.loc[df['price'] > 100  , 'price_val'] = 'Above 100'
```

```python
from  sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()

df.country=lb.fit_transform(df.country)
```

```python
df.wine_quality=lb.fit_transform(df.wine_quality)
```

```python
df.winery=lb.fit_transform(df.winery)
```

```python
df.variety=lb.fit_transform(df.variety)
```

```python
df.price_val=lb.fit_transform(df.price_val)
```

```python
df.province=lb.fit_transform(df.province)
```

```python
df.region_1=lb.fit_transform(df.region_1)
```

```python
df.drop(['price','points'],axis=1,inplace=True)
```

```python
df
```

```python
x=df.drop(['review_description','variety'],axis=1)
y=df.variety
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.15,random_state=42)
```

```python
xtrain
```

```python
len(ytrain.unique())
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(xtrain,ytrain)
kpred=knn.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,kpred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(max_depth=50,min_samples_split=2,criterion="gini")
dt.fit(xtrain,ytrain)
dpred=dt.predict(xtest)
accuracy_score(ytest,dpred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(max_depth=50,n_estimators=150,criterion='gini')
rf.fit(xtrain,ytrain)
rpred=rf.predict(xtest)
accuracy_score(ytest,rpred)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier(n_estimators=150,max_leaves=0)
xgb.fit(xtrain,ytrain)
xpred=xgb.predict(xtest)
accuracy_score(ytest,xpred)
```

```python
from sklearn.ensemble import VotingClassifier
meta_model = VotingClassifier(
    estimators=[
        ('dt', dt),
        ('rf', rf),
        ('xgb', xgb)
    ],
    voting='hard'  # Use 'soft' for classifiers that can predict probabilities
)
meta_features = np.column_stack((dpred, rpred, xpred))
meta_model.fit(meta_features, ytest)
# Generate meta-features for the test set using the base models
meta_features_test = np.column_stack((
    dt.predict(xtest),
    rf.predict(xtest),
    xgb.predict(xtest)
))

# Make predictions using the meta-model
stacked_predictions = meta_model.predict(meta_features_test)


```

```python
accuracy_score(ytest,stacked_predictions)
```
